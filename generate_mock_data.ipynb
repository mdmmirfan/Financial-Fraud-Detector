```python
import pandas as pd
import numpy as np
import os
import glob

# Corrected paths based on your VS Code workspace
real_data_path = 'data/' 
mock_data_path = 'mock_data/'

os.makedirs(mock_data_path, exist_ok=True)

# Find all CSVs in your dataset folder
for file in glob.glob(real_data_path + '*.csv'):
    # Read just the column names and data types from the real file
    real_df = pd.read_csv(file, nrows=5) 
    columns = real_df.columns
    
    # Create an empty dataframe with the same columns
    mock_df = pd.DataFrame(columns=columns)
    
    num_mock_rows = 50 # Create 50 fake rows
    
    for col in columns:
        # If it's a numeric column, generate random numbers
        if pd.api.types.is_numeric_dtype(real_df[col]):
            mock_df[col] = np.random.randint(0, 1000, size=num_mock_rows)
        # If it's text/string, generate random letters/words
        else:
            mock_df[col] = ['fake_data_' + str(i) for i in range(num_mock_rows)]
            
    # Save the fake data into the new mock_data folder
    filename = os.path.basename(file)
    mock_df.to_csv(mock_data_path + filename, index=False)
    print(f"✅ Created 50 rows of mock data for {filename}")
```
